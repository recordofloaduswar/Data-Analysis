```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv('C:/Projects/Data Analysis/Spotify 2014 - 2022/charts.csv', low_memory = False)
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
df = df[df['streams'].str.contains("\(") == False]
df.shape
```

```python
df = df.astype({'date': 'datetime64[ns]', 'streams': int})
df.dtypes
```

```python
find_null = df.isnull().sum()
find_null
```

```python
null_data = df[df.isnull().any(axis=1)]
null_data = null_data.sort_values('artist_genres')
null_data.head()
```

```python
null_data.tail()
```

```python
df = df[df['name'].notna()]

find_null2 = df.isnull().sum()
find_null2
```

```python
df.shape
```

```python
df.artists = df.artists.str.replace('[', '', regex = True).replace(']', '', regex = True).replace("'", '', regex = True)
df.artist_genres = df.artist_genres.str.replace('[', '', regex = True).replace(']', '', regex = True).replace("'", '', regex = True)
df.head()
```
